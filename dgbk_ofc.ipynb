# Importanto e analisando dados
Etree element: biblioteca Python que permite analisar e navegar em um documento XML

```python
import zipfile
import xml.etree.ElementTree as ET
import pandas as pd

zip_file = "drugbank_all_full_database.xml.zip"

with zipfile.ZipFile(zip_file, 'r') as z:
    # Listando os arquivos
    xml_files = [file_name for file_name in z.namelist() if file_name.endswith(".xml")]

    if not xml_files:
        raise ValueError("Nenhum arquivo XML encontrado")

    xml_file_name = xml_files[0]

    with z.open(xml_file_name) as xml_file:
        tree = ET.parse(xml_file)
        root = tree.getroot()

data = []
columns = set()

for elem in root:
    row_data = {}
    for subelem in elem:
        columns.add(subelem.tag)
        row_data[subelem.tag] = subelem.text

    # Preenchendo as colunas ausentes com valores nulos
    row_data = {col: row_data.get(col, None) for col in columns}
    data.append(row_data)

df = pd.DataFrame(data, columns=sorted(columns))

print(df.head())

csv_path = '/content/resultado.csv'
df.to_csv(csv_path, index=False)

print(f'CSV salvo em: {csv_path}')

df.rename(columns=(''))
```
